## Missing Values

```python
#Missing values stats
data.isna().sum()
```

Since the main purpose of this work is Bokeh visualization library demo, we could reduce our dataset's size without loosing consistency by omitting the fileds with the most missing values.

```python
#Reduced size dataset
data_reduced = data[['country', 'year', 'status', 'life expectancy',
       'measles', 'bmi', 'under-five deaths', 'polio', 'hiv/aids']]
```

```python
#Reduced size dataset's missing values stats
data_reduced.isna().sum()
```

```python
#Life expectancy
data_reduced[data_reduced['life expectancy'].isna()==True]
```

```python
#BMI
data_reduced[data_reduced['bmi'].isna()==True]
```

```python
#Polio
data_reduced[data_reduced['polio'].isna()==True]
```

As one can see, missing values do not pose any significant threat to the idea of future dataset's use for vizulaization purposes, hence we could drop the data entries with missed fields entirely.

```python
#Dropping entries with missing values and creating a cleaned dataset
data_clean = data_reduced.dropna()

#Checking the clean dataset
data_clean.isna().sum()
```
